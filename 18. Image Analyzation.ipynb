## Image Analyzation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Using Histogram

```python
img = np.zeros((200,200), np.uint8)

# First Rectangle
cv2.rectangle(img, (0,100), (200,200), (255), -1)
# Second Rectangle
cv2.rectangle(img, (0,50), (50,100), (127), -1)

# Parameters (src, channel, mask, histsize, range)
hist = cv2.calcHist([img], [0], None, [256], [0,256])

plt.plot(hist)
plt.show()

cv2.imshow('res',img)
cv2.waitKey(0)
cv2.destroyAllWindows()

print('First rectangle makes most of pixels white and few pixels of gray are made from our 2nd rectangle')
```

### 2. Color Image

```python
img = cv2.imread('Images/Image1.jpg')
img = cv2.resize(img, (400,600))

b, g, r = cv2.split(img)

plt.hist(b.ravel(), 256, [0,256])
plt.hist(g.ravel(), 256, [0,256])
plt.hist(r.ravel(), 256, [0,256])

plt.title('Colored Image')
plt.show()

cv2.imshow('Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 3. Image Pixel Equalization

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Equalized nearby pixels to same value
equ = cv2.equalizeHist(gray)
cv2.imshow('Equalized Image',equ)

nhist = cv2.calcHist([equ], [0], None, [256], [0,256])
plt.plot(nhist)
plt.title('Equalization')
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 4. CLAHE Technique
- clipLimit: The contrast limit for limiting amplification. This parameter determines the maximum allowed contrast enhancement. Higher values result in stronger contrast enhancement. Setting it too high may produce an unnatural appearance.
- tileGridSize defines the size of the contextual region used for histogram equalization. It specifies the number of tiles in the grid. A larger grid size provides finer control but can lead to slower processing.

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))

cl = clahe.apply(gray)
cv2.imshow('Clahe', cl)

chist = cv2.calcHist([cl], [0], None, [256], [0,256])
plt.plot(chist)
plt.title('CLAHE')
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 5. Image Visualization with all Techniques

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('Images/Image1.jpg', 0)

# Perform histogram equalization
equ = cv2.equalizeHist(img)

# Perform CLAHE
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
clahe_img = clahe.apply(img)

# Calculate histograms
hist_orig = cv2.calcHist([img], [0], None, [256], [0, 256])
hist_equ = cv2.calcHist([equ], [0], None, [256], [0, 256])
hist_clahe = cv2.calcHist([clahe_img], [0], None, [256], [0, 256])

# Plot histograms
plt.figure(figsize=(30, 30))

# no of rows '2', no of columns '3', its position '1'
plt.subplot(231)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(232)
plt.imshow(cv2.cvtColor(equ, cv2.COLOR_BGR2RGB))
plt.title('Equalization Image')

plt.subplot(233)
plt.imshow(cv2.cvtColor(clahe_img, cv2.COLOR_BGR2RGB))
plt.title('Clahe Image')

plt.subplot(234)
plt.plot(hist_orig, color='black')
plt.title('Original Image Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.subplot(235)
plt.plot(hist_equ, color='blue')
plt.title('Histogram Equalization')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.subplot(236)
plt.plot(hist_clahe, color='green')
plt.title('CLAHE')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
