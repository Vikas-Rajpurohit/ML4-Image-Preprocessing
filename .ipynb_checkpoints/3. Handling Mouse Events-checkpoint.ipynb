## 1. Handling Mouse Events

```python
import cv2
import numpy as np
```

```python
# Creating Blank black image
img = np.zeros((512,512,3), np.uint8)
```

```python
def draw(event, x, y, flags, params):
    if event == cv2.EVENT_LBUTTONDBLCLK:
        cv2.circle(img, (x,y), 50, (125,0,255), 5)
    if event == cv2.EVENT_RBUTTONDBLCLK:
        cv2.rectangle(img, (x,y), (x+100,y+100), (125,125,255), 5)
```

## 2. Color Picker

```python
def mouse_event(event, x, y, flags, params):
    if event == cv2.EVENT_LBUTTONDOWN:
        cord = ". "+str(x)+", "+str(y)
        cv2.putText(img, cord, (x,y), cv2.FONT_HERSHEY_PLAIN, 1 ,(155,125,100), 2)
    if event == cv2.EVENT_RBUTTONDOWN:
        b = img[y, x, 0]
        g = img[y, x, 1]     
        r = img[y, x, 2]   
        
        color = ". "+str(b)+", "+str(g)+", "+str(r)
        cv2.putText(img, color, (x,y), cv2.FONT_HERSHEY_PLAIN, 1, (152, 255, 130), 2)
```

```python
cv2.namedWindow(winname = 'res')

# define Callback function on Mouse Event
cv2.setMouseCallback('res',mouse_event)

while True:
    cv2.imshow('res',img)
    
    # 27 -> Escape key
    if cv2.waitKey(1) & 0xFF == 27:
        break
        
cv2.destroyAllWindows() 
```

```python
import matplotlib.pyplot as plt

image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```
