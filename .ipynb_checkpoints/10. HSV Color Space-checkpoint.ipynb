## 1. HSV Color Space

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Images\hsv.png')

res = plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.show(res)
```

```python
# HSV Masking Detect

img = cv2.imread('Images/Image1.jpg')

while True:
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    # BGR Format
    u_v = np.array([30,170,255]) # Upper Value of Yellow
    l_v = np.array([21,126,178]) # Lower Value of Yellow
    
    # Creating a Binary Mask 
    # that identifies the pixels within the specified HSV range.
    # The mask is constructed by setting pixel values to 255 (white) 
    # if they fall within the specified range and 0 (black) otherwise
    mask = cv2.inRange(hsv, l_v, u_v)
    
    # Filter with Mask
    res = cv2.bitwise_and(img, img, mask=mask)
    res = cv2.resize(res, (400,600))
     
    cv2.imshow('Result',res)
    
    if cv2.waitKey(0) & 0xFF == 27:
        break
        
cv2.destroyAllWindows()
```
